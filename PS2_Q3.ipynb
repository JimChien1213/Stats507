```python
import pandas as pd
```

```python
# Question 3
# a
url = [
    'https://wwwn.cdc.gov/Nchs/Nhanes/2011-2012/DEMO_G.XPT',
    'https://wwwn.cdc.gov/Nchs/Nhanes/2013-2014/DEMO_H.XPT',
    'https://wwwn.cdc.gov/Nchs/Nhanes/2015-2016/DEMO_I.XPT',
    'https://wwwn.cdc.gov/Nchs/Nhanes/2017-2018/DEMO_J.XPT',
]
year = ['2011-2012',
       '2013-2014',
       '2015-2016',
       '2017-2018']
init = pd.DataFrame([])

for i in range(len(year)):
    data = pd.DataFrame(pd.read_sas(url[i]))
    data.insert(0, column='year', value=year[i])
    init = pd.concat([init, data])
column_1 = init[['SEQN', 'RIDAGEYR', 'RIDRETH3', 'DMDEDUC2', 'DMDMARTL',
                 'RIDSTATR', 'SDMVPSU', 'SDMVSTRA']].fillna(value=-1).astype(int)
column_2 = init[['WTMEC2YR', 'WTINT2YR']]
column_3 = init[['year']]

column_1['RIDRETH3'].replace({1:'Mexican American', 2:'Other Hispanic', 3:'Non-Hispanic White',
                              4: 'Non-Hispanic Black', 6: 'Non-Hispanic Asian',
                              7: 'Other Race - Including Multi-Racial'}, inplace=True)
column_1['DMDEDUC2'].replace({1:'Less than 9th grade', 2:'9-11th grade (Includes 12th grade with no diploma)', 
                              3:'High school graduate/GED or equivalent', 4: 'Some college or AA degree', 
                              5: 'College graduate or above', 7:'Refused', 9: "Don't Know"}, inplace=True)
column_1['DMDMARTL'].replace({1:'Married', 2:'Widowed', 3:'Divorced', 4: 'Separated', 5: 'Never married', 
                              6:'Living with partner', 77:'Refused', 99: "Don't Know"}, inplace=True)
column_1['RIDSTATR'].replace({1:'Interviewed only', 2:'Both interviewed and MEC examined'}, inplace=True)

comb = pd.concat([column_1, column_2], axis=1).fillna(value=-1)
comb.replace({-1: 'Missing'}, inplace=True)
comb_rename = comb.rename(columns = {'SEQN':'ids', 'RIDAGEYR':'age', 'RIDRETH3':'race and ethnicity',
                                     'DMDEDUC2':'education', 'DMDMARTL':'marital status', 
                                     'RIDSTATR':'interview/examination status', 
                                     'SDMVPSU':'masked variance pseudo-PSU', 
                                     'SDMVSTRA':'masked variance pseudo-stratum',
                                     'WTMEC2YR':'full sample 2 year MEC exam weight', 
                                     'WTINT2YR':'full sample 2 year interview weight'})

final = pd.concat([column_3, comb_rename], axis=1)
final.columns = final.columns.map(lambda x:x.lower())
final.reset_index(inplace=True, drop=True)
final.to_pickle('demographic.pickle')
final_pickle = pd.read_pickle('demographic.pickle')

# test
final
```

```python
# Question 3
# b
url = [
    'https://wwwn.cdc.gov/Nchs/Nhanes/2011-2012/OHXDEN_G.XPT',
    'https://wwwn.cdc.gov/Nchs/Nhanes/2013-2014/OHXDEN_H.XPT',
    'https://wwwn.cdc.gov/Nchs/Nhanes/2015-2016/OHXDEN_I.XPT',
    'https://wwwn.cdc.gov/Nchs/Nhanes/2017-2018/OHXDEN_J.XPT',
]
year = ['2011-2012',
       '2013-2014',
       '2015-2016',
       '2017-2018']
init = pd.DataFrame([])

for i in range(len(year)):
    data = pd.DataFrame(pd.read_sas(url[i]))
    data.insert(0, column='year', value=year[i])
    init = pd.concat([init, data])
column_1 = init[['SEQN']].astype(int)
column_2 = init[['OHDDESTS']]
column_3 = init.filter(regex='CTC$')
column_4 = init[['year']]

dict_ohx = {b'A':'Primary tooth with a restored surface condition',
           b'D':'Sound primary tooth',
           b'E':'Missing due to dental disease',
           b'F':'Permanent tooth with a restored surface condition',
           b'J':'Permanent root tip is present but no restorative replacement is present',
           b'K':'Primary tooth with a dental carious surface condition',
           b'M':'Missing due to other causes',
           b'P':'Missing due to dental disease but replaced by a removable restoration',
           b'Q':'Missing due to other causes but replaced by a removable restoration',
           b'R':'Missing due to dental disease but replaced by a fixed restoration',
           b'S':'Sound permanent tooth',
           b'T':'Permanent root tip is present but a restorative replacement is present',
           b'U':'Unerupted',
           b'X':'Missing due to other causes but replaced by a fixed restoration',
           b'Y':'Tooth present, condition cannot be assessed',
           b'Z':'Permanent tooth with a dental carious surface condition'
           }
dict_ctc = {'OHX01CTC':'Upper right 3rd molar (3M)',
            'OHX02CTC':'Upper right 2nd molar (2M)',
            'OHX03CTC':'Upper right 1st molar (1M)',
            'OHX04CTC':'Upper right 2nd bicuspid/2nd primary molar (2B)',
            'OHX05CTC':'Upper right 1st bicuspid/1st primary molar (1B)',
            'OHX06CTC':'Upper right cuspid (C)',
            'OHX07CTC':'Upper right lateral incisor (LI)',
            'OHX08CTC':'Upper right central incisor (CI)',
            'OHX09CTC':'Upper left central incisor (CI)',
            'OHX10CTC':'Upper left lateral incisor (LI)',
            'OHX11CTC':'Upper left cuspid (C)',
            'OHX12CTC':'Upper left 1st bicuspid/1st primary molar (1B)',
            'OHX13CTC':'Upper left 2nd bicuspid/2nd primary molar (2B)',
            'OHX14CTC':'Upper left 1st molar (1M)',
            'OHX15CTC':'Upper left 2nd molar (2M)',
            'OHX16CTC':'Upper left 3rd molar (3M)',
            'OHX17CTC':'Lower left 3rd molar (3M)',
            'OHX18CTC':'Lower left 2nd molar (2M)',
            'OHX19CTC':'Lower left 1st molar (1M)',
            'OHX20CTC':'Lower left 2nd bicuspid/2nd primary molar (2B)',
            'OHX21CTC':'Lower left 1st bicuspid/1st primary molar (1B)',
            'OHX22CTC':'Lower left cuspid (C)',
            'OHX23CTC':'Lower left lateral incisor (LI)',
            'OHX24CTC':'Lower left central incisor (CI)',
            'OHX25CTC':'Lower right central incisor (CI)',
            'OHX26CTC':'Lower right lateral incisor (LI)',
            'OHX27CTC':'Lower right cuspid (C)',
            'OHX28CTC':'Lower right 1st bicuspid/1st primary molar (1B)',
            'OHX29CTC':'Lower right 2nd bicuspid/2nd primary molar (2B)',
            'OHX30CTC':'Lower right 1st molar (1M)',
            'OHX31CTC':'Lower right 2nd molar (2M)',
            'OHX32CTC':'Lower right 3rd molar (3M)'
           }

column_3_replace = column_3.replace(dict_ohx)
column_3_rename = column_3_replace.rename(columns=dict_ctc)
comb = pd.concat([column_4, column_1, column_2, column_3_rename], axis=1)
comb_2 = comb.fillna(value=-1)
comb_2['OHDDESTS'].replace({1:'Complete', 2:'Partial', 3:'Not Done'}, inplace=True)
comb_2.replace({-1: 'Missing'}, inplace=True)

final = comb_2.rename(columns={'SEQN':'ids', 'OHDDESTS':'overall oral health exam status'})
final.columns = final.columns.map(lambda x:x.lower())
final.reset_index(inplace=True, drop=True)
final.to_pickle('dentition.pickle')
final_pickle = pd.read_pickle('dentition.pickle')

# test
final
```
