```python
# Question 1

import math
import time
import statistics as st
import pandas as pd

# a
def fib_rec(n):
    if n == 0 or n == 1:
        return n
    else:
        return (fib_rec(n - 1) + fib_rec(n - 2))
# b
def fib_for(n):
    F_n = []
    F_n.append(0)
    F_n.append(1)
    for i in range(2, n + 1):
        F_n.append(F_n[i - 1] + F_n[i - 2])
    return F_n[n]
# c
def fib_whl(n):
    F_n = []
    i = 0
    F1, F2 = 0, 1
    if n == 0:
        F_n = 0
    elif n == 1:
        F_n = 1
    else:
        while i < n - 1:
            F_n = F1 + F2
            F1, F2 = F2, F_n
            i += 1
    return F_n
# d
def fib_rnd(n):
    phi = (1 + 5 ** 0.5) / 2
    F_n = round((phi ** n) / 5 ** 0.5)
    return F_n
# e
def fib_flr(n):
    phi = (1 + 5 ** 0.5) / 2
    F_n = math.floor((phi ** n) / 5 ** 0.5 + 0.5)
    return F_n

def test():
    test_num = [7, 11, 13]
    for i in range(3):
        n = test_num[i]
        print('fib_rec(' + str(n) + ')=' + str(fib_rec(n)))
        print('fib_for(' + str(n) + ')=' + str(fib_for(n)))
        print('fib_whl(' + str(n) + ')=' + str(fib_whl(n)))
        print('fib_rnd(' + str(n) + ')=' + str(fib_rnd(n)))
        print('fib_flr(' + str(n) + ')=' + str(fib_flr(n)))
    print()
test()

# f
t1, t2, t3, t4, t5 = [], [], [], [], []
for n in range(10, 36, 5):
    for run in range(10):
        start = time.time()
        fib_rec(n)
        end1 = time.time()
        fib_for(n)
        end2 = time.time()
        fib_whl(n)
        end3 = time.time()
        fib_rnd(n)
        end4 = time.time()
        fib_flr(n)
        end5 = time.time()
        t1.append(round(end1 - start, 3))
        t2.append(round(end2 - end1, 7))
        t3.append(round(end3 - end2, 7))
        t4.append(round(end4 - end3, 7))
        t5.append(round(end5 - end4, 7))
t1_md = [st.median(t1[i:i + 10]) for i in range(0, len(t1), 10)]
t2_md = [st.median(t2[i:i + 10]) for i in range(0, len(t2), 10)]
t3_md = [st.median(t3[i:i + 10]) for i in range(0, len(t3), 10)]
t4_md = [st.median(t4[i:i + 10]) for i in range(0, len(t4), 10)]
t5_md = [st.median(t5[i:i + 10]) for i in range(0, len(t5), 10)]
df =  pd.DataFrame({'Recursive': t1_md,
                   'For Loop': t2_md,
                   'While Loop': t3_md,
                   'Rounding': t4_md,
                   'Truncation': t5_md},
                   columns = ['Recursive', 'For Loop', 'While Loop', 'Rounding', 'Truncation'],
                   index = ['n=10', 'n=15', 'n=20', 'n=25', 'n=30', 'n=35'])
print('Medain of Calculation Time (sec)')
print(df)
```

```python
# Question 3

import numpy as np
import scipy.stats as st
import pandas as pd

# a
def normal(data, level, ci_format):
    try:
        dec = 3
        a = np.array(data)
        n = len(a)
        se = st.sem(a)
        x = round(np.mean(a), dec)
        z = st.t.ppf((1 + level) / 2, n-1)
        lwr, upr = round(x - z * se, dec), round(x + z * se, dec)
        form = '{x}[{confidence}% CI:({lwr},{upr})]'
        dict = {'x': x, 'lwr': lwr, 'upr': upr, 'confidence': level*100}
        if ci_format == None:
            output = dict
        else:
            output = form.format_map(dict)
        return output
    except Exception as e:
        print("Please input the right format!!!")
# Testing for a
data = [1, 2, 2, -1, 5, 2]
level = 0.95
ci_format = 'string'
# print(normal(data, level, ci_format))

# b
def binomial(data, level, ci_format, method):
    try:
        dec = 3
        a = np.array(data)
        n = len(a)
        alpha = 1 - level
        mean = round(np.mean(a), dec)
        # In this case, assume that 1 is the successes 
        x = (a == 1).sum()
        if method == 'normal_app':
            phat = round(x / n, dec)
            if min(n * phat, n * (1 - phat)) > 12:
                z = st.t.ppf((1 + level) / 2, n - 1)
                lwr, upr = round(phat - z * (phat * (1 - phat) / n) ** 0.5, dec), round(phat + z * (phat * (1-phat) / n) ** 0.5, dec)
            else:
                print('either (n x phat) or (n x (1 - phat) should > 12')

        elif method == 'clopper':
            lwr, upr = round(st.beta.ppf(alpha / 2, x, n - x + 1), dec), round(st.beta.ppf(1 - alpha / 2, x + 1, n-x), dec)
        elif method == 'jeffrey':
            lwr, upr = round(max(st.beta.ppf(alpha / 2, x + 0.5, n - x + 0.5),0), dec), round(min(st.beta.ppf(1 - alpha / 2, x + 0.5, n-x + 0.5),1), dec)
        else:
            z = st.t.ppf((1 + level) / 2, n - 1)
            nt = n + z ** 2
            pt = round( (x + z ** 2) / 2 / nt, dec)
            lwr, upr = round(pt - z * (pt * (1 - pt) / n) ** 0.5, dec), round(pt + z * (pt * (1 - pt) / n) ** 0.5, dec)
        dict = {'x': mean, 'lwr': lwr, 'upr': upr, 'confidence': level * 100}        
        form = '{x}[{confidence}% CI:({lwr},{upr})]'
        if ci_format == None:
            output = dict
        else:
            output = form.format_map(dict)
        return output  
    except Exception as e:
        print("Please input the right format!!!")

# c
data = np.hstack((np.ones(42), np.zeros(48)))
level = [0.90, 0.95, 0.99]
method = ['normal_app', 'clopper', 'jeffrey', 'agresti']
ci_format = 'string'
output_90, output_95, output_99 = [], [], []
for i in range(0, 3):
    if i == 0:
        output_90.append(normal(data, level[i], ci_format))
    elif i == 1:
        output_95.append(normal(data, level[i], ci_format))
    else:
        output_99.append(normal(data, level[i], ci_format))
    for ii in range(0, 4):
        if i == 0:
            output_90.append(binomial(data, level[i], ci_format, method[ii]))
        elif i == 1:
            output_95.append(binomial(data, level[i], ci_format, method[ii]))
        else:
            output_99.append(binomial(data, level[i], ci_format, method[ii]))
df =  pd.DataFrame({'Level = 90%': output_90,
                   'Level = 95%': output_95,
                   'Level = 99%': output_99},
                   columns = ['Level = 90%', 'Level = 95%', 'Level = 99%'],
                   index = ['Normal Theory', 'Normal Approximation', 'Clopper-Pearson Interval',
                            'Jeffrey’s Interval', 'Agresti-Coull Interval'])
print(df)
print('')
print('Agesti-Coull Interval produces the interval with the smallest width')
```
